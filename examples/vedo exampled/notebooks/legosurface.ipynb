```python
"""Represent a volume as lego blocks (voxels). Colors correspond to the volume's scalar."""
from vedo import dataurl, settings, Volume, Plotter

vol = Volume(dataurl+'embryo.tif') # load Volume
vol.print()

# show lego blocks whose value is between vmin and vmax
lego = vol.legosurface(vmin=65).cmap('seismic').add_scalarbar()

plt = Plotter(axes=1, bg='wheat', bg2='lb', size=(900,600))
plt.show(lego, viewup='z', zoom=1.75)
```

```python
plt.close()
```
